# Challenge 2 - Data Cleaning and Transformation

If your conclusion from the previous challenge is the data need cleaning/transformation, do it in the cells below. However, if your conclusion is the data need not be cleaned or transformed, feel free to skip this challenge. But if you do choose the latter, please provide rationale.

```python
# Your code here

#detectar y eliminar OUTLIERS

# Calcular el rango intercuartílico (IQR)
Q1 = data.quantile(0.25)  # Primer cuartil
Q3 = data.quantile(0.75)  # Tercer cuartil
IQR = Q3 - Q1

# Determinar los límites inferior y superior
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar los outliers
cleaned_data = data[~((data < lower_bound) | (data > upper_bound)).any(axis=1)]

# Mostrar información antes y después
print(f"Datos originales: {data.shape}")
print(f"Datos después de limpiar outliers: {cleaned_data.shape}")
```

```python
#NORMALIZAR VARIABLES con MIN MAX SCALER

from sklearn.preprocessing import MinMaxScaler

# Crear el escalador
scaler = MinMaxScaler()

# Aplicar el escalado a las columnas numéricas
columns_to_scale = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
scaled_data = scaler.fit_transform(cleaned_data[columns_to_scale])

# Crear un nuevo DataFrame con los datos escalados
normalized_data = cleaned_data.copy()
normalized_data[columns_to_scale] = scaled_data

# Verificar las primeras filas del dataset normalizado
print(normalized_data.head())

```

# Challenge 3 - Data Preprocessing

One problem with the dataset is the value ranges are remarkably different across various categories (e.g. `Fresh` and `Grocery` compared to `Detergents_Paper` and `Delicassen`). If you made this observation in the first challenge, you've done a great job! This means you not only completed the bonus questions in the previous Supervised Learning lab but also researched deep into [*feature scaling*](https://en.wikipedia.org/wiki/Feature_scaling). Keep on the good work!

Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

```python
# Your import here:

from sklearn.preprocessing import StandardScaler

# Crear el escalador
scaler = StandardScaler()

# Escalar los datos numéricos
columns_to_scale = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
customers_scale = scaler.fit_transform(cleaned_data[columns_to_scale])

# Verificar la data escalada
print(customers_scale[:5])  # Mostrar las primeras 5 filas escaladas

```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
from sklearn.cluster import KMeans

# Your code here:

# Iniciar el modelo de K-Means
kmeans = KMeans(n_clusters=3, random_state=42)  # Puedes ajustar el número de clusters según sea necesario

# Ajustar el modelo a los datos escalados
kmeans.fit(customers_scale)

# Obtener los labels asignados por K-Means
customers = cleaned_data.copy()  # Crear una copia del dataset limpio original
customers['labels'] = kmeans.labels_

# Ver las primeras filas del dataset con las etiquetas asignadas
print(customers.head())
```

### Looking to the elbow we can choose 2 like the correct number of clusters

```python
from sklearn.cluster import KMeans

# Entrenar el modelo con 2 clústeres
kmeans_2 = KMeans(n_clusters=2, random_state=42).fit(customers_scale)

# Predecir las etiquetas (opcional, ya están en labels_)
labels = kmeans_2.predict(customers_scale)

# Convertir las etiquetas en una lista
clusters = kmeans_2.labels_.tolist()
```

```python
# Asignar las etiquetas al dataset principal
customers['Label'] = clusters
```

Count the values in `labels`.

```python
# Your code here:
# Asignar las etiquetas al dataset principal
customers['Label'] = clusters

# Contar los valores en la columna "Label"
label_counts = customers['Label'].value_counts()

print(customers.head())  # Ver las primeras filas con las etiquetas
print("\nCount of labels:\n", label_counts)
```

# Challenge 5 - Data Clustering with DBSCAN

Now let's cluster the data using DBSCAN. Use `DBSCAN(eps=0.5)` to initiate the model, then fit your scaled data. In the data returned from the `.fit` method, assign the `labels_` back to `customers['labels_DBSCAN']`. Now your original data have two labels, one from K-Means and the other from DBSCAN.

```python
from sklearn.cluster import DBSCAN 

# Your code here
# Iniciar el modelo DBSCAN con eps=0.5
dbscan = DBSCAN(eps=0.5, min_samples=5)

# Ajustar el modelo a los datos escalados
dbscan.fit(customers_scale)

# Asignar las etiquetas generadas por DBSCAN al dataset
customers['labels_DBSCAN'] = dbscan.labels_

# Verificar las primeras filas del dataset con las etiquetas DBSCAN
print(customers.head())

# Contar los valores en la columna "labels_DBSCAN"
dbscan_label_counts = customers['labels_DBSCAN'].value_counts()
print("\nCount of DBSCAN labels:\n", dbscan_label_counts)
```

# Bonus Challenge 3 - Changing DBSCAN `eps` and `min_samples`

Experiment changing the `eps` and `min_samples` params for DBSCAN. See how the results differ with scatter plot visualization.

```python
# Your code here
from sklearn.cluster import DBSCAN
import seaborn as sns
import matplotlib.pyplot as plt

# Función para graficar
def plot_dbscan(x, y, eps_val, min_samples_val, title):
    # Aplicar DBSCAN con nuevos parámetros
    dbscan = DBSCAN(eps=eps_val, min_samples=min_samples_val)
    dbscan.fit(customers_scale)

    # Asignar las nuevas etiquetas de DBSCAN
    customers['labels_DBSCAN'] = dbscan.labels_

    # Graficar los resultados con nuevas etiquetas
    sns.scatterplot(x=x, y=y, hue='labels_DBSCAN', data=customers, palette='viridis', alpha=0.6)
    plt.title(f'{title} (eps={eps_val}, min_samples={min_samples_val})')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.legend(title='DBSCAN Labels')
    plt.show()

# Experimentar con diferentes valores de eps y min_samples
# Cambiar eps y min_samples para ver cómo afectan la clasificación
plot_dbscan('Detergents_Paper', 'Milk', eps_val=0.3, min_samples_val=5, title="Detergents Paper vs Milk")
plot_dbscan('Detergents_Paper', 'Milk', eps_val=0.7, min_samples_val=5, title="Detergents Paper vs Milk")
plot_dbscan('Detergents_Paper', 'Milk', eps_val=0.5, min_samples_val=10, title="Detergents Paper vs Milk")
```

**Your comment here**

- eps mas bajo (0.3) genera mas ruido, con menos clusters definidos
- min_samples mas alto (10) reduce los clusters, pero los hace mas grandes y cohesivos
- eps mas alto (0.7) aumenta el tamaño de los cllusters pero tambien puede incluir más puntos extraños
